```python
import pandas as pd
import numpy as np
import geopandas as gpd
import dask
```

```python
walkability_df = gpd.read_file('/Volumes/Rohun Backups/Datasets/Natl_WI_SHP/Natl_WI.shp')
```

```python
walkability_df.shape
```

```python
walkability_df.head()
```

```python
walkability_df = pd.DataFrame(walkability_df)
```

```python
walkability_df = walkability_df.loc[(walkability_df.SFIPS == '36') & 
                                    (walkability_df.CFIPS.isin(['005', '047','061', '081', '085']))]
```

```python
walkability_df.head()
```

```python
df_trial = walkability_df.groupby(['TRFIPS', 'CFIPS']).mean().reset_index()
```

```python
df_trial.head()
```

```python
df_trial['tract_code'] = df_trial['CFIPS'] + df_trial['TRFIPS']
```

```python
df_trial.head()
```

```python
nta_tracts = pd.read_excel('/Volumes/Rohun Backups/Datasets/nyc2010census_tabulation_equiv.xlsx')
```

```python
nta_tracts.columns = nta_tracts.iloc[0]
nta_tracts = nta_tracts.reindex(nta_tracts.index.drop(0))
nta_tracts.head()
```

```python
nta_tracts['full_code'] = nta_tracts['2010 Census Bureau FIPS County Code'] + nta_tracts['2010 Census Tract']
```

```python
df_trial = df_trial.merge(nta_tracts, how = 'left', left_on = 'tract_code', right_on = 'full_code')
```

```python
df_trial.head()
```

```python
df_trial.columns
```

```python
df_for_mlc = df_trial.drop(['TRFIPS', 'CFIPS', 'CBSA', 'COUNTHU10', 'TOTPOP10', 'HH', 'WORKERS',
               'AC_TOT', 'AC_WATER', 'AC_LAND', 'AC_UNPR','Shape_Leng', 'Shape_Area', 'tract_code', 'Borough',
               '2010 Census Bureau FIPS County Code', '2010 NYC Borough Code', '2010 Census Tract', 'PUMA',
               'Name', 'full_code'], axis = 1)
df_for_int = df_trial.drop(['TRFIPS', 'CFIPS', 'CBSA', 'COUNTHU10', 'TOTPOP10', 'HH', 'WORKERS',
               'AC_TOT', 'AC_WATER', 'AC_LAND', 'AC_UNPR','Shape_Leng', 'Shape_Area', 'tract_code', 'Borough',
               '2010 Census Bureau FIPS County Code', '2010 NYC Borough Code', '2010 Census Tract', 'PUMA',
               'Name'], axis = 1)
df_for_mlc.head()
```

```python
df_for_mlc.to_csv('walkability_index_mlc.csv')
df_for_int.to_csv('walkability_index_int.csv')
```
